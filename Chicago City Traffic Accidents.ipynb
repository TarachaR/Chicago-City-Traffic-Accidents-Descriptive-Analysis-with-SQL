```
import pandas as pd

traffic = pd.read_csv('traffic_accidents.csv', index_col=False)

%sql DROP TABLE if EXISTS traffic

%sql PERSIST traffic;
```

```
traffic.tail()
```

```
%%sql
SELECT * FROM traffic LIMIT 5;
```

```
#  Display a list of 5 makes that had the highest number of crashes
%%sql
SELECT DISTINCT make 
FROM traffic
WHERE unit_no > 3 LIMIT 5
```

```
# 1. Most personal vehicles involved in crashes were travelling towards which direction? 

%%sql
SELECT travel_direction, MAX(unit_no)
FROM traffic
WHERE vehicle_use = 'PERSONAL'
```

```
# 2. Most personal vehicles involved in crashes were travelling towards which direction

%%sql
SELECT travel_direction FROM traffic WHERE vehicle_use = 'personal' ;
```

```
# 3. How many passengers were involved in personal cars were in accidents?

%%sql
SELECT SUM(num_passengers) FROM traffic WHERE vehicle_use = 'PERSONAL';
```

```
# 4. How many passengers were involved in accidents while travelling in personal Sport Utility vehicles

%%sql
SELECT SUM(num_passengers) 
FROM traffic 
WHERE vehicle_use = 'PERSONAL'
AND vehicle_type = 'SPORT UTILITY VEHICLE (SUV)';
```

```
# 5. Compute the total number of Mercedez-Benz vehicles involved in accidents while being parked.

%%sql 
SELECT COUNT(make) 
FROM traffic
WHERE make = 'MERCEDES-BENZ'
AND maneuver = 'PARKED';
```

```
# 6. Compute the total occupant count per vehicle count.
%%sql
SELECT SUM(occupant_cnt)
FROM traffic
```

```
# 7. Sort the number of passengers in descending order by 
# travel direction taking into account travel direction. 
# Return travel direction and no. of passengers.

%%sql
SELECT num_passengers, travel_direction
FROM traffic
WHERE num_passengers IS NOT NULL
ORDER BY num_passengers DESC LIMIT 5;
```

```
# 8. Select top 5 vehicle models involved in the accident with no defect, 
# order them by make and show their manoeuvre. 

%%sql
SELECT model, maneuver
FROM traffic
WHERE vehicle_defect = 'NONE' 
ORDER BY make LIMIT 10;
```

```
# 9. Which ridesharing vehicles in an accident had the most number of passengers? 
# Return model, and no of passengers.

%%sql
SELECT model, num_passengers
FROM traffic
WHERE vehicle_use = 'RIDESHARE SERVICE'
ORDER BY num_passengers DESC;
```
